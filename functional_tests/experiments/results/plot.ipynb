```python
import pyreadr

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cdr = pyreadr.read_r('discovery_cdr_1000s_2r.rdata')["causal_decision_rules"]
cdr.replace("NaN","0", inplace=True)
cdr.iloc[:,1:] = cdr.iloc[:,1:].astype(float)
cdr
```

```python
em = pyreadr.read_r('discovery_em_1000s_2r.rdata')["effect_modifiers"]
em.replace("NaN","0", inplace=True)
em.iloc[:,1:] = em.iloc[:,1:].astype(float)
em
```

```python
fig, ax = plt.subplots(1,3,figsize=(10, 3), layout='constrained')
fig.suptitle('Discovery - Effect Modifiers')

sns.lineplot(data=em, x="effect_size", y="IoU", hue="method", ax=ax[0])
ax[0].set_xlabel('Effect Size') 
ax[0].set_ylabel('IoU') 
ax[0].set_ylim((0,1))
ax[0].legend(loc='upper left');  

sns.lineplot(data=em, x="effect_size", y="Precision", hue="method", ax=ax[1])
ax[1].set_xlabel('Effect Size')  
ax[1].set_ylabel('Precision')  
ax[1].set_ylim((0,1))
ax[1].legend(loc='upper left');  

sns.lineplot(data=em, x="effect_size", y="Recall", hue="method", ax=ax[2])
ax[2].set_xlabel('Effect Size') 
ax[2].set_ylabel('Recall')  
ax[2].set_ylim((0,1))
ax[2].legend(loc='upper left');  
```
